```python
import matplotlib.pyplot as plt
from sklearn import datasets, metrics
from sklearn.model_selection import train_test_split

digits = datasets.load_digits()
plt.imshow(digits.images[0], cmap=plt.cm.gray_r, interpolation='nearest')
    
#이미지 전처리 과정 2차원데이터를 1차원 데이터로 reshape 1차원 벡터로 펴기 
n_samples = len(digits.images)
data = digits.images.reshape((n_samples, -1))

from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=6) #knn머신러닝 기법 가져오기

X_train, X_test, y_train, y_test = train_test_split(
    data, digits.target, test_size=0.2)

knn.fit(X_train, y_train)

#테스트 데이터로 예측해보기
y_pred = knn.predict(X_test)
#정확도 계산
scores = metrics.accuracy_score(y_test, y_pred)
print(scores)
#이미지를 출력하기 위해 평탄화된 이미지를 다시 8x8 형상으로 만든다
plt.imshow(X_test[10].reshape(8,8), cmap=plt.cm.gray_r, interpolation='nearest')

y_pred = knn.predict([X_test[10]]) #입력은 항상 2차원 행렬이어야함
print(y_pred)
```

```python
import cv2 as cv

image = cv.imread('num.png', cv.IMREAD_GRAYSCALE)

image = cv.resize(image, (8, 8))

image = image.astype('float32')

image = image.reshape(1, 64)

image = 255-image

image /= 255.0

plt.imshow(image.reshape(8,8), cmap=plt.cm.gray_r, interpolation='nearest')
y_pred = knn.predict(image)
print(y_pred)
```
