```python
import numpy as np
```

```python
#시그모이드 함수

def actf(x):
    return 1 / (1 + np.exp(-x))
#시그모이드함수 미분치
def actf_deriv(x):
    return x * (1 - x)
```

```python
inputs, hiddens, outputs = 2, 2, 1
learning_rate = 0.2
```

```python
X = np.array([[0,0], [0, 1], [1, 0], [1, 1]])
T = np.array([[0], [1], [1], [0]])
```

```python
W1 = np.array([[0.10, 0.20], 
               [0.30, 0.40]])
W2 = np.array([[0.50], [0.60]])

B1 = np.array([0.1, 0.2])
B2 = np.array([0.3])
```

```python
def predict(x):
    layer0 = x
    Z1 = np.dot(layer0, W1) + B1
    layer1 = actf(Z1)
    Z2 = np.dot(layer1, W2) + B2
    layer2 = actf(Z2)
    return layer0, layer1, layer2
```

```python
def test():
    for x, y in zip(X, T):
        x = np.reshape(x, (1, -1))
        layer0, layer1, layer2 = predict(x)
        print(x, y, layer2)
```

```python
test()
```
